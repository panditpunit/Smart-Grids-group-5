```python

demand = pd.read_excel('data.xlsx')
plt.style.use('ggplot')
plt.figure(figsize=(15,5))
plt.plot(demand['Hour'], demand['Demand'], linewidth=0.9, color= 'r', label='Demand [p.u.]')
plt.plot(demand['Hour'], demand['PV'], linewidth=0.9, color= 'g', label='PV generation [p.u.]')
plt.plot(demand['Hour'], demand['Wind'], linewidth=0.9, color= 'b', label='Wind generation [p.u.]')
plt.title('Demand vs PV and Wind generation [p.u.]')
plt.xlabel('Hour')
plt.ylabel('Power [p.u.]')
plt.legend()
plt.xlim([0, 175])
plt.ylim([0, 1])

plt.savefig('Demand_Gen_Plot.png')
plt.show()
demand
```

```python
demand = pd.read_excel('day_data.xlsx')
plt.style.use('ggplot')
plt.figure(figsize=(15,5))

plt.plot(demand['Hour'], demand['Demand'], linewidth=0.9, color= 'r', label='Demand [p.u.]')
plt.plot(demand['Hour'], demand['PV'], linewidth=0.9, color= 'g', label='PV generation [p.u.]')
plt.plot(demand['Hour'], demand['Wind'], linewidth=0.9, color= 'b', label='Wind generation [p.u.]')

plt.title('Daily demand [p.u.]')
plt.xlabel('Hour')
plt.ylabel('Power demanded [p.u.]')
plt.legend()
plt.xlim([0, 23])
plt.ylim([0, 1])

plt.savefig('Demand_Plot.png')
plt.show()
```

## Data Importing

#### Import 1st iteration results

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


results=pd.read_excel('iteration_results.xlsx')
results=results.drop(results.columns[[0]], axis=1, inplace=False)

voltage=results.drop(results.iloc[:,17:],axis = 1, inplace=False)


angle=results.drop(results.iloc[:,34:],axis = 1, inplace=False)
angle=angle.drop(angle.iloc[:,:17],axis = 1, inplace=False)

lines=results.drop(results.iloc[:,42:],axis = 1, inplace=False)
lines=lines.drop(lines.iloc[:,:34],axis = 1, inplace=False)

trafos=results.drop(results.iloc[:,:42],axis = 1, inplace=False)

```

```python


## VOLTAGE PLOT FOR ALL BUSES - WHITE BACKGROUND

plt.style.use('ggplot')
plt.figure(figsize=(15,5))

ax=plt.gca()
ax.set_facecolor("white")

plt.plot(voltage)
plt.title('Bus voltages in p.u.')
plt.xlabel('Hour')
plt.ylabel('Voltage')
plt.legend(['V0','V1','V2','V3','V4','V5','V6','V7','V8','V9','V10','V11','V12','V13','V14','V15','V16'],loc='upper left', title='Labels',labelcolor="black", facecolor="white")

plt.grid(color="black",linestyle='-', linewidth=0.2)
plt.tight_layout(pad=1.08)


plt.savefig('Voltages_1.png')


```
